```python
import pandas as pd
#path1是去重后的
#path2是abandon的
path1 = '/mnt/vepfs/lingxin/Pretrain-data/wulindong/juben_crawl/dudmp_data_exist/weixin_page.2018_special_token_3000.jsonl'
path2 = '/mnt/vepfs/lingxin/Pretrain-data/wulindong/juben_crawl/dudmp_data_abandon/weixin_page.2018_special_token_3000.jsonl'
df1 = pd.read_json(path1, lines=True)
df2 = pd.read_json(path2, lines=True)
df = pd.concat([df1, df2])
gf = df.groupby('__cluster__')
gf_d = gf.filter(lambda x: len(x)>1)
len(gf_d)
gs = gf_d.groupby('__cluster__').groups
gs
```

```python
import numpy as np
group_names = list(gs.keys())
random_group_name = np.random.choice(group_names)
random_group_indices = gs[random_group_name]
random_group = gf_d.loc[random_group_indices]
# random_group
texts = random_group['text']
for i in range(len(texts)):
    print(texts.iloc[i])
    print("_______________________________________________________________________________")
```

```python
text = texts.iloc[2]
```

```python
import re
import re

def extract_chinese(text):
    # 使用正则表达式匹配中文字符
    chinese_pattern = re.compile('[^\u4e00-\u9fa5]')
    # 将非中文字符替换为 ""
    chinese_text = chinese_pattern.sub('', text)
    # 去掉中括号包裹的内容
    chinese_text = re.sub(r'\[.*?\]', '', chinese_text)
    # 返回处理后的中文字符串
    return chinese_text
extract_chinese(text)
```
