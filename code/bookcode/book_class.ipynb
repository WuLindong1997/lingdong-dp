```python
# 定义字典
import matplotlib.pyplot as plt
import json
import numpy as np

```

```python
with open('/mnt/vepfs/lingxin/Pretrain-data/wulindong/code/bookcode/length.json','r',encoding='utf-8')as file:
    d = json.loads(file.read())

# 对字典按照值进行排序
keys = sorted(map(int, d.keys()))
# 初始化一个长度为最大键值加1的列表
data_list = [0] * (keys[-1] + 1)

# 遍历键值对，将值存储到对应的下标位置
for k, v in d.items():
    data_list[int(k)] = v
```

```python
data_list = data_list[:38560]
```

```python
data_list
```

```python
sum_before_10 = sum(data_list[:10])
idx_before_10 = [0]
# 计算每隔50个元素的和
sum_every_50 = []
idx_every_50 = []
for i in range(10, len(data_list), 20):
    idx_every_50.append(i)
    sum_every_50.append(sum(data_list[i:i+20]))
```

```python
result_list = [sum_before_10] + sum_every_50
result_idx = idx_before_10+idx_every_50
```

```python
result_list
```

```python
plt.pie(result_list, labels=result_idx)
plt.title('length distribution')
plt.show()
```
