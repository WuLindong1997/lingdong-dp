```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

# 假设data是一个字典，包含每个关键字及其出现次数
data = {'每日穿搭': 735713,
 '好物分享': 647802,
 '日常穿搭': 619224,
 '香奈儿Chanel': 536656,
 '每秒都值得记录': 514591,
 '我的日常': 427720,
 '穿搭': 456953,
 '周末去哪儿': 414827,
 'ootd每日穿搭': 301172,
 '不懂就问有问必答': 300403}
# 将关键字及其出现次数转成TF-IDF特征向量
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(list(data.keys()))

# 构建K-Means模型并进行聚类
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# 输出每个关键字所属的类别
labels = kmeans.labels_
for i, (j,label) in enumerate(zip(data.keys(),labels)):
    print(f"{j}\t{label}")
```
