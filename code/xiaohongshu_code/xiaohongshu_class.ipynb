```python
# 定义字典
import matplotlib.pyplot as plt
import json
import numpy as np

```

```python
with open('/mnt/vepfs/lingxin/pretrain/wulindong/xiaohongshu_classes.json','r',encoding='utf-8')as file:
    d = json.loads(file.read())

# 对字典按照值进行排序
del d['##']
sorted_values = sorted(d.values(), reverse=True)
```

```python
d
```

```python
# 计算前100个值的和和所有值的和
list_step = []
list_percentage = []
length = len(sorted_values)
max_num = sum(sorted_values)
for i in range(1,length,10000):
    print(i)
    top100_sum = sum(sorted_values[:i])
    list_step.append(i)
    list_percentage.append(top100_sum/max_num)
    print(top100_sum/max_num)
```

```python
length
```

```python
list_percentage[:10]
```

```python
plt.plot(list_step, list_percentage)
plt.xlabel('Sort Top')
plt.ylabel('Percentage')
plt.xticks(np.linspace(0, length, num=18, endpoint=True, dtype=int),rotation = 40)
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```
